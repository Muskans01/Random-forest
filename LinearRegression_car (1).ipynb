```python
def univariate_categorical(data,var):
    miss=data[var].isnull().sum()
    uncat_cnt=data[var].nunique()
    uncat_list=data[var].unique()
    
    h=pd.DataFrame(data[var].value_counts()).reset_index()
    h.columns=[var, "Count"]
    
    k=pd.DataFrame(data[var].value_counts(normalize=True)).reset_index()
    k.columns=[var, "percentage"]
    k["percentage"]=(round(k["percentage"]*100, 2)).astype("str")+str("%")
    final=pd.merge(h,k, on=var, how="inner")
    
    print(f"Missing counts: {miss}\n")
    print(f"Total Unique count of categories: {uncat_cnt} \n")
    print(f"Unique category list: \n{uncat_list} \n")
    
    plt.figure(figsize=(8,6))
    sns.countplot(data=data, x=var)
    plt.xticks(rotation=90)
    plt.show()
    return final
```

```python
univariate_categorical(data=mydata, var="Manufacturer")
```

```python
def univariate_numerical(data, var):
    results={"Missing":data[var].isnull().sum(),
            "min": data[var].min(),
            "max":data[var].max(),
            "sum":data[var].sum(),
            "average":data[var].mean(),
            "standard Deviation":data[var].std(),
            "Variance":data[var].var(),
            "skewness":data[var].skew(),
            "Kurtosis":data[var].kurt(),
            "25th percentile":data[var].quantile(.25),
            "50th percentile":data[var].quantile(.5),
            "75th percentile":data[var].quantile(.75)}
    plt.figure(figsize=(12,4))
    plt.subplot(121)
    sns.histplot(data[var])
    plt.subplot(122)
    sns.boxplot(data=data, y=var)
    plt.show()
    return results
```

```python
def missings(data):
    misss=pd.DataFrame(data.isnull().sum(), columns=["Missing_count"]).reset_index()
    
    misss["Missing_percentage"]=round(misss["Missing_count"]/data.shape[0]*100,2).astype("str")+str("%")
    
    miss0=misss[misss["Missing_count"]>0].sort_values(by="Missing_count", ascending=False)
    return miss0
```
